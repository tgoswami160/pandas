# Merge



```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2=pd.DataFrame({"a":[1,2,3,4],"c":[21,22,23,24]})
pd.merge(var1,var2,on="a")
```

```python
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2=pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var2,var1,on="a")
```

```python
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2=pd.DataFrame({"a":[1,2,3,5],"c":[21,22,23,24]})
pd.merge(var1,var2,how="outer",indicator=True)#how=inner,left,right,outer
```

```python
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2=pd.DataFrame({"a":[1,2,3,4],"b":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,12,13,14])
pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
d2=pd.DataFrame({"a":[1,2,3,5],"b":[21,22,23,24]})

pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
d2=pd.DataFrame({"a":[1,2,5],"b":[21,22,23]})

pd.concat([d1,d2],axis=1,join="inner")#inner-intersection,outer=union
```

```python
d11=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
d12=pd.DataFrame({"a":[1,2,5],"b":[21,22,23]})

pd.concat([d11,d12],axis=0,keys=["d11","d12"])
```

```python
d1=pd.DataFrame({"a":[1,2,3,4]})
d2=pd.DataFrame({"a":[1,2,5],"b":[21,22,23]})

pd.concat([d1,d2])
```

# Join

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,22,22,11]},index=["a","b","c","d"])
var2=pd.DataFrame({"c":[10,20],"d":[21,29]},index=["a","b"])

var1.join(var2)
```

```python
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,22,22,11]},index=["a","b","c","d"])
var2=pd.DataFrame({"c":[10,20],"d":[21,29]},index=["a","b"])

var2.join(var1,how="outer",rsuffix="_12")#how=inner,outer,left,right
```

# Append

```python
var1=pd.DataFrame({"a":[1,2,3,4],"b":[11,22,22,11]},index=["a","b","c","d"])
var2=pd.DataFrame({"c":[10,20],"b":[21,29]},index=["a","b"])

var1.append(var2,ignore_index=True)
```

# pivot()

```python
var=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","a","a","b","b"],"eng":[10,12,14,16,18,12],"maths":[17,18,19,13,14,16]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

```python
var=pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","a","a","b","b"],"eng":[10,12,14,16,18,12],"maths":[17,18,19,13,14,16]})
var
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean")
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum",margins="True")
```
